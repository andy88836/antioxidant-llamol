```python
from rdkit import Chem
from rdkit.Chem import AllChem
import pandas as pd
import glob

data = []

# 假设您的 .xyz 文件位于名为 'qm9_xyz' 的文件夹中
for file in glob.glob('qm9_xyz/*.xyz'):
    with open(file, 'r') as f:
        lines = f.readlines()
        num_atoms = int(lines[0])
        properties = lines[1].strip().split()
        mol_block = ''.join(lines[2:num_atoms+2])
        mol = Chem.MolFromXYZBlock(mol_block)
        if mol:
            props = {
                'property1': properties[0],
                'property2': properties[1],
                # 添加更多属性
            }
            data.append(props)

df = pd.DataFrame(data)
df.to_csv('qm9.csv', index=False)
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem import RDConfig
import os
import sys
sys.path.append(os.path.join(RDConfig.RDContribDir, 'SA_Score'))
# now you can import sascore!
import sascorer


def calculate_sascore(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        sascore = sascorer.calculateScore(mol)
        return sascore
    except:
        return None

# 读取CSV文件
df = pd.read_csv(r'D:\MolGen-main\AODB_antio_compounds.csv')

# 计算SAScore并添加新列
df['SAScore'] = df['SMILES'].apply(calculate_sascore)

# 保存到新的CSV文件
df.to_csv('AODB_sascore.csv', index=False)

print("SAScore has been calculated and added as a new column.")
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Crippen

# 假设你的数据集是一个 CSV 文件
# 数据集中的 SMILES 列为 "smiles"
df = pd.read_csv("AODB_sascore.csv")

# 定义计算 logP 值的函数
def calculate_logp(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        return Crippen.MolLogP(mol)
    else:
        return None

# 应用函数计算 logP 值
df["logP"] = df["SMILES"].apply(calculate_logp)

# 保存结果到新的 CSV 文件
df.to_csv("AODB_logP.csv", index=False)

print(df.head())
```

```python
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
def calcContextHBD(smiles):
    hbd = []
    for smi in smiles:
        mol = Chem.MolFromSmiles(smi)
        hb = rdMolDescriptors.CalcNumHBD(mol)
        hbd.append(hb)
    return np.array(hbd)

smiles = ["CCO", "CCN", "CCC"]  # 示例 SMILES 列表
new_context = {}  # 新的上下文字典

# 调用函数并赋值
new_context['hbd'] = calcContextHBD(smiles)
```
